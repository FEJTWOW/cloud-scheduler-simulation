```python
for file in usa_list:
    loaded_file = pickle.load(open(f'{pwd}/data/predictions/{file}', 'rb'))
    splitted = file.split('.pkl')
    n_splitted = splitted[0].split('_usa')
    for data, state_name in zip(loaded_file, state_list):
        df  = pd.DataFrame.from_dict(data, orient ='index')
        df = df.reset_index()
        df.columns = ['datetime', 'carbon_per_MWh']
        df.to_csv(f'{pwd}/data/USA/predictions/{state_name}_{n_splitted[0]}.csv', index=False)
```
