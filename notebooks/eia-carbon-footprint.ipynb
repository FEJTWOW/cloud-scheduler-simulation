CO_2 emission by fuel (based on 2020 data):  
Coal : 2.23 pounds/kWh  
Natural Gas: 0.91 pounds/kWh  
Petroleum: 2.13 pounds/kWh  

other sources are considered to be carbon neutral.

```python
df = pd.read_csv("../data/USA/Midwest.csv", index_col=0)
df
```

```python
# sum of energy 
df["MWh"] = df.sum(axis=1)
df
```

```python
is_col = "COL" in df.columns
is_ng = "NG" in df.columns
is_oil = "OIL" in df.columns
is_col, is_ng, is_oil
```

```python
def calculate_carbon_footpring(row):
    cf = 0
    if is_col:
        cf += row.COL * 2.23
    if is_ng:
        cf += row.NG * 0.91
    if is_oil:
        cf += row.OIL * 2.13
    return cf * 1000 / 2.205
```

```python
df.assign(
    carbon_footprint = lambda r: calculate_carbon_footpring(r),
    carbon_per_MWh = lambda r: r.carbon_footprint / r.MWh
)
```
