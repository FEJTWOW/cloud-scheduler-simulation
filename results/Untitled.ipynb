```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
naive = pd.read_csv('naive_dnn.csv')['0'].values
opt = pd.read_csv('opt_dnn.csv')['0'].values
```

```python
dnaive = naive[-1:]-naive[:-1]
dopt = opt[-1:]-opt[:-1]
```

```python
plt.plot(naive,label='naive_cumulated')
plt.plot(opt,label='optimized_cumulated')
plt.ylabel('Skumulowana emisja dwutlenku węgla')
plt.xlabel('Czas [godzina]')
plt.legend()
plt.show()
```

```python
plt.plot(dnaive,label='naive')
plt.plot(dopt,label='optimized')
plt.ylabel('Emisja dwutlenku węgla')
plt.xlabel('Czas [godzina]')
plt.legend()
plt.show()
```
